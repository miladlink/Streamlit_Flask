```python
import pytorch_lightning as pl
from pytorch_lightning.metrics.functional import accuracy

import matplotlib.pyplot as plt
import seaborn as sns

import os
import torch
import torch.nn as nn
from torchvision import transforms
from sklearn.metrics import confusion_matrix, classification_report

import mlflow.pytorch
from mlflow.tracking import MlflowClient

from model import *
from data import *
```

```python
imgs_dir = '/content/stage_2_train_images'
lbls_dir = '/content/train_labels.csv'
```

```python
# transformation on dataset
train_tfms = transforms.Compose([
    transforms.RandomHorizontalFlip (),
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize (mean = [0.485, 0.456, 0.406], 
                          std = [0.229, 0.224, 0.225])
])

valid_tfms =  transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize (mean = [0.485, 0.456, 0.406], 
                          std = [0.229, 0.224, 0.225])
])
```

```python
# read labels
labels = read_labels (lbls_dir)
# shuffle split and store images and labels in varibles
train_paths, valid_paths, train_lbls, valid_lbls = shuffle_split_path (imgs_dir, labels, seed = 3)
# create train and valid loader
train_dl, valid_dl = get_loaders (
    train_paths, valid_paths, train_lbls, valid_lbls,
      16, train_tfms, valid_tfms)
```

```python
class TorchClassifier (pl.LightningModule):
    def __init__ (self, model_name):
        super (TorchClassifier, self).__init__ ()
        
        self.backbone = torchModel (model_name)
        self.loss_fn = nn.CrossEntropyLoss ()
          
    def forward (self, x):
        return self.backbone (x)

    def configure_optimizers (self):
        optimizer = optimizer = torch.optim.Adam (self.parameters (), lr = 0.0002)
        return optimizer

    def training_step (self, batch, batch_idx):
        images, labels = batch
        outputs = self (images)
        loss = self.loss_fn (outputs, labels)
        acc = accuracy (outputs, labels)
        self.log ('train_acc', acc)
        self.log ('train_loss', loss, on_epoch=True)
        pbar = {'acc': acc}
        return {'loss': loss, 'progress_bar': pbar}

    def validation_step (self, batch, batch_idx):
        images, labels = batch
        outputs = self (images)
        val_loss = self.loss_fn (outputs, labels)
        val_acc = accuracy (outputs, labels)
        self.log ('val_acc', val_acc)
        self.log ('val_loss', val_loss, on_epoch=True)
        pbar = {'val_acc': val_acc}
        return {'val_loss': val_loss, 'progress_bar': pbar}

    def validation_epoch_end (self, val_step_outputs):
        avg_val_loss = torch.tensor ([x ['val_loss'] for x in val_step_outputs]).mean ()
        avg_val_acc = torch.tensor ([x ['progress_bar']['val_acc'] for x in val_step_outputs]).mean ()

        pbar = {'avg_val_acc': avg_val_acc}
        return {'avg_val_loss': avg_val_loss, 'progress_bar': pbar}

    def test_step(self, batch, batch_idx):
        images, labels = batch
        outputs = self (images)
        test_loss = self.loss_fn (outputs, labels)
        test_acc = accuracy (outputs, labels)
        self.log ('test_acc', test_acc)
        self.log ('test_loss', test_loss, on_epoch=True)
        pbar = {'test_acc': test_acc}
        return {'test_loss': test_loss, 'progress_bar': pbar}

    def test_epoch_end (self, test_step_outputs):
        avg_test_loss = torch.tensor ([x ['test_loss'] for x in test_step_outputs]).mean ()
        avg_test_acc = torch.tensor ([x ['progress_bar']['test_acc'] for x in test_step_outputs]).mean ()

        pbar = {'avg_test_acc': avg_test_acc}
        return {'avg_test_loss': avg_test_loss, 'progress_bar': pbar}


    def predict_step (self, batch, batch_idx, dataloader_idx = None):
        images, labels = batch
        return self (images)
```

# ResNet50

* train full of model

```python
model = TorchClassifier ('resnet50')
```

```python
trainer = pl.Trainer (progress_bar_refresh_rate = 20, max_epochs = 5, gpus = 1)
```

```python
# Train the model
with mlflow.start_run () as run:
    trainer.fit (model, train_dataloaders = train_dl, val_dataloaders = valid_dl)
```

```python
trainer.validate (model, valid_dl)
```

```python
trainer.validate (ckpt_path = 'best')
```

```python
trainer.test (model, valid_dl)
```

```python
trainer.test (ckpt_path = 'best', dataloaders = valid_dl)
```

```python
prediction = trainer.predict (model, valid_dl)
```

```python
prediction [0].shape
```

```python
def print_auto_logged_info (r):

    tags = {k: v for k, v in r.data.tags.items () if not k.startswith ("mlflow.")}
    artifacts = [f.path for f in MlflowClient().list_artifacts (r.info.run_id, "model")]
    print("run_id: {}".format (r.info.run_id))
    print("artifacts: {}".format (artifacts))
    print("params: {}".format (r.data.params))
    print("metrics: {}".format (r.data.metrics))
    print("tags: {}".format (tags))
```

```python
y_test, y_pred = [], []
for imgs, lbls in valid_dl:
    outs = model (imgs)
    _, preds = torch.max (outs, dim = 1)
    y_test += lbls.tolist ()
    y_pred += preds.tolist ()
print ('done!')
```

```python
cm = confusion_matrix (y_test, y_pred, normalize = 'true')
cm_df = pd.DataFrame (cm)
plt.figure (figsize = (12, 10))
plt.title ('Confusion Matrix')
sns.heatmap (cm_df, annot = True, cmap = 'Blues', square = True)
```

```python
print_auto_logged_info (mlflow.get_run (run_id = run.info.run_id))
```
